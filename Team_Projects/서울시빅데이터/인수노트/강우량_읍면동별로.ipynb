```python
import pandas as pd
```

```python
Data=pd.read_csv("./강우량_종힙.csv")
```

```python
Data.drop([0,1],axis=0,inplace=True)
Data.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
Data.head(3)
```

```python
Data.info()
```

```python
Data[["시우량","일년누계","최대우량"]]=Data[["시우량","일년누계","최대우량"]].applymap(float)
```

```python
use_Data=Data[Data["시우량"]!=0]
use_Data
```

time_series로 변환

```python
use_Data["송신지_자료수집_시각"]=use_Data["송신지_자료수집_시각"].apply(lambda x:pd.to_datetime(x))
```

초 단위 제거

```python
use_Data["송신지_자료수집_시각"]=use_Data["송신지_자료수집_시각"].apply(lambda x:x.replace(second=0))
use_Data
```

살펴봐야할것
- 1. 최대우량
- 2. 시우량
- 3. 송신지_자료수집_시각

```python
use_Data.sort_values(["구청명","송신지_자료수집_시각"],ascending=True,inplace=True)
```

```python
use_Data.drop(columns=["강우량_코드","구청_코드","수신서버_저장_시각"],inplace=True)
```

```python
use_Data["date"]=use_Data["송신지_자료수집_시각"].apply(lambda x:f"{x.year}.{x.month}.{x.day}")
use_Data["hour"]=use_Data["송신지_자료수집_시각"].apply(lambda x:x.hour)
use_Data["min"]=use_Data["송신지_자료수집_시각"].apply(lambda x:x.minute)
```

구마다 비가 온 날짜가 다름

```python
groups=use_Data.groupby(["구청명"]).groups
groups_idx=[x[1] for x in groups.items()]
```

```python
test=use_Data.loc[groups_idx[0]]
test["diff"]=test["송신지_자료수집_시각"].diff()
```
