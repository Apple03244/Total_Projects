```python
import pandas as pd
```

```python
배수등급=pd.read_csv("../동별_인구_고도_배수등급_공시지가_결합.csv")
```

```python
배수등급["SOILDRA"]
```

```python
배수등급표=pd.DataFrame(배수등급.groupby(["GU_NAME","ADM_NM"])["SOILDRA"].value_counts())
```

```python
전체count=pd.DataFrame(배수등급.groupby(["ADM_NM"])["SOILDRA"].count())
```

```python
전체count
```

```python
배수등급비율=pd.merge(배수등급표.reset_index(),전체count,left_on="ADM_NM",right_on="ADM_NM",how="left")
```

```python
idx1=배수등급비율["SOILDRA_x"]=="양호"
idx2=배수등급비율["SOILDRA_x"]=="매우양호"
idx3=배수등급비율["SOILDRA_x"]=="약간양호"
```

```python
양호count=pd.DataFrame(배수등급비율[idx1|idx2|idx3].groupby(["ADM_NM"])["count"].sum())
```

```python
양호count
```

```python
merge_비율=pd.merge(양호count,전체count,left_index=True,right_index=True)
```

```python
merge_비율["양호비율"]=merge_비율["count"]/merge_비율["SOILDRA"]
```

```python
merge_비율.reset_index()
```

```python
배수등급표.reset_index()
```

```python
test_df=pd.merge(merge_비율.reset_index(),배수등급표.reset_index()[["ADM_NM","GU_NAME"]],how="inner")
```

```python
test_df=test_df.drop_duplicates(["ADM_NM"])
```

```python
test_df.set_index(["GU_NAME","ADM_NM"]).rename(columns={'count':"양호개수","SOILDRA":"배수전체개수"})
```
