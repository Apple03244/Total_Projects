```python
def recluster(x):
    if x==0:
        return "2"
    elif x==1:
        return "0"
    else:
        return "1"
```

```python
Data['Cluster']=Data["Cluster"].apply(recluster)
Data["Cluster"]=Data["Cluster"].astype("int")
```
