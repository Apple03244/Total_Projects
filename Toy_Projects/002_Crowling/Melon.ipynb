```python
def transtime(x):
    return "".join(str(x).split(" ")[0].split("-"))
```

#### Selenium

```python
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import bs4
import pandas as pd
import time
import datetime as dt
```

```python
chromedriver_autoinstaller.install()
driver=webdriver.Chrome()
driver.implicitly_wait(5)
```

```python
days_dic={}
days=pd.date_range(pd.to_datetime("20230102"),dt.datetime.now(),freq="7D")
for day in days:
    end=day+pd.Timedelta(days=6)
    if not transtime(end)[4:-2] in days_dic:
        days_dic[transtime(end)[4:-2]]=[list(map(transtime,[day,end]))]
    else:
        days_dic[transtime(end)[4:-2]].append(list(map(transtime,[day,end])))
```

```python
urls={}
for month in days_dic:
    if month=="07":
        break
    for d in days_dic[month]:
        urls[d[0]]=f'https://www.melon.com/chart/search/list.htm?chartType=WE&age=2020&year=2023&mon={month}&day={d[0]}%5E{d[1]}&classCd=DM0000&startDay={d[0]}&endDay={d[1]}&moved=Y'
```

```python
def drive_get(url):
    driver.get('https://www.melon.com/')
    driver.get(url)
    time.sleep(1)
    print("변경완료")        
```

```python
Melon={}
for day in urls:
    drive_get(urls[day])
    soup=bs4.BeautifulSoup(driver.page_source)
    ranking_data=[]
    for i in range(100):
        insert_data={}
        row=soup.select("#chartListObj tr")[i]
        insert_data["ranking"]=row.select(".rank")[0].text.strip()
        insert_data["title"]=row.select(".ellipsis.rank01 a")[0].text.strip()
        insert_data["artist"]=row.select(".ellipsis.rank02 a")[0].text.strip()
        ranking_data.append(insert_data)
    Melon[day]=ranking_data
```
