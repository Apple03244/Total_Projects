```python
#행정구역 설정 함수
def trans(x):
    for i in 행정구역.keys():
        try:
            if i in x:
                return 행정구역[i]
        except:
            break
```

```python
data["행정구역"]=data["지번주소"].apply(lambda x:trans(x))
data.loc[data["행정구역"].isna(),"행정구역"]=data[data["행정구역"].isna()].apply(lambda x:trans(x))
```
