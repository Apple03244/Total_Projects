## 문제가 되는 column들 확인을 위한 pvalue 확인

```python
import scipy as sp
import itertools
```

### column 쌍 만들기

```python
comb=[x for x in itertools.combinations(ordinal.columns,r=2)]
```

### 데이터 프레임으로 확인하기

```python
data=[]
for x in comb:
    col=[x]
    col.extend(sp.stats.spearmanr(ordinal[list(x)]))
    data.append(col)
result_df=pd.DataFrame(data=data)
result_df.rename(columns=dict(zip(result_df.columns,["columns","value","pvalue"])),inplace=True)
```

```python
# 상관계수의 값에 따른 분류
def indepen(x):
    if abs(x)>=0.2:
        return False
    else:
        return True
```

```python
result_df["result"]=result_df["value"].apply(indepen)
result_df
```

#### 결과확인
